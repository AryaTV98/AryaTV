```python
df5=df.copy()
```

```python
df=pd.get_dummies(df)
df
```

```python
x=df.drop(['price'],axis=1)
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

```python

from sklearn.neighbors import KNeighborsRegressor
Classifier=KNeighborsRegressor(n_neighbors=5)
Classifier.fit(x_train,y_train)
y_pred=Classifier.predict(x_test)
print(y_pred)
print(Classifier.predict([[1,748,2,4,0,1,4,1,19.75,40]]))
```

```python
from sklearn.metrics import mean_absolute_error
error=mean_absolute_error(y_test,y_pred)
error
```

```python
from sklearn.metrics import mean_absolute_percentage_error
per=mean_absolute_percentage_error(y_test,y_pred)
per
```

```python
from sklearn.metrics import mean_squared_error
squ=mean_squared_error(y_test,y_pred)
squ
```

```python
sqrt=np.sqrt(squ)
sqrt
```

```python
#r2 score
from sklearn .metrics import r2_score
r2=r2_score(y_test,y_pred)
r2
```

```python
output=pd.DataFrame({'price_actual':y_test,'price_pred':y_pred})
result=df5.merge(output,left_index=True,right_index=True)
```

```python
result.sample(10)
```

```python
df7=result.groupby(['airline','source_city','destination_city',],as_index=False)['price_pred'].mean()
df7
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x='days_left',y='price_actual',data=result,color='red')
sns.lineplot(x='days_left',y='price_pred',data=result,color='blue')
plt.title('Days left Vs actual and predicted Price',fontsize=15)
plt.xlabel('Days left',fontsize=15)
plt.ylabel('Actual and Predicted Price',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20,15))
sns.jointplot(x='price_pred',y='price_actual',data=result,kind='reg',color='red')
plt.title('Actual vs Predicted Price',fontsize=15)
plt.xlabel('predicted Price',fontsize=15)
plt.ylabel('Actual price',fontsize=15)
plt.show()
```
